```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine
import urllib # Needed for properly formatting the URL string

# --- Connection Parameters ---
SERVER = 'LENOVO\\SQLEXPRESS' 
DATABASE = 'Laptop_Pricing'
DRIVER = 'ODBC Driver 17 for SQL Server' # Use 17 or 18, based on your install

# 1. Properly format the connection string components for a SQLAlchemy URL
# This encodes the driver name (e.g., replaces spaces with '+')
params = urllib.parse.quote_plus(
    f"DRIVER={DRIVER};"
    f"SERVER={SERVER};"
    f"DATABASE={DATABASE};"
    f"Trusted_Connection=yes;"
)

# 2. Create the SQLAlchemy Engine URL
# Syntax: mssql+pyodbc:///?odbc_connect=<encoded_pyodbc_string>
# This tells SQLAlchemy to use the MSSQL dialect via the pyodbc connector.
sqlalchemy_url = f"mssql+pyodbc:///?odbc_connect={params}"

# 3. Create the SQLAlchemy Engine
engine = create_engine(sqlalchemy_url)

# --- Load Data to SQL Server using the Engine ---
try:
    # Use the new SQLAlchemy engine object for the 'con' argument
    df1.to_sql(
        name="Laptops_Data", 
        con=engine,  # <--- Now passing the SQLAlchemy Engine!
        if_exists='replace',    
        index=False             
    )
    print("Data successfully loaded to SQL Server using SQLAlchemy Engine.")

except Exception as e:
    print(f"An error occurred during data loading: {e}")

finally:
    # 4. Close the engine connection pool
    if 'engine' in locals():
        engine.dispose()
        print("SQLAlchemy Engine disposed.")
```
